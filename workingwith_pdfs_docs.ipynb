```python
# glob() method
# the programmers require to traverse through a list of files at some location, mostly having a
# specific pattern. Python’s glob module has several functions that can help in listing files 
# under a specified folder. We may filter them based on extensions, or with a particular string 
# as a portion of the filename.
# All the methods of Glob module follow the Unix-style pattern matching mechanism and rules.
# However, it doesn’t allow expanding the tilde (~) and environment variables.
```

# list of files with patterns

# wildcards asterick (*) matches zero or more characters in a segment of a name. For example, dir/*.

```python
# The fallowing code will show the contents of covid directory
import glob
for name in glob.glob('covid/*'):
    print(name)
```

```python
import glob
for name in glob.glob('test_glob_mainfolder/*'):
    print(name)
```

```python
# The first case above lists the subdirectory name explicitly, while the second case depends
# on a wildcard to find the directory.
import glob

print('Named with wildcard:')
for name in glob.glob('test_glob_mainfolder/*/*'):
    print('\t', name)
```

# Zip files

```python
# contents of zipped file

from zipfile import ZipFile
with ZipFile('pythonloops.zip', 'r') as zipObj:
   # Get list of files names in zip
   listOfiles = zipObj.namelist()
   # Iterate over the list of file names in given list & print them
   for elem in listOfiles:
        print(elem)
```

```python
# Detailed Info about of files in a zip files
```

```python
with ZipFile('python loops.zip', 'r') as zipObj:
   # Get list of ZipInfo objects
   listOfiles = zipObj.infolist()
   # Iterate of over the list of ZipInfo objects & access members of the object
   for elem in listOfiles:
        print(elem.filename, ' : ', elem.file_size, ' : ', elem.date_time, ' : ', elem.compress_size)
```

```python
# Print the details of ZIP archive to std.out using ZipFile.printdir()

with ZipFile('python loops.zip', 'r') as Zobj:
    Zobj.printdir()
```

```python
# importing the libraries
from bs4 import BeautifulSoup
import requests

url="https://americanliterature.com/short-stories-for-children"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
print(soup.prettify()) # print the parsed data of html
```

```python
#Scrapy , Textract , poppler-utils , antiword
print(soup.title)
```

```python
#To get the text without the HTML tags
print(soup.title.text)
```

```python
Now, let's get all the links in the page along with its attributes, such as href, title, 
and its inner Text.

for link in soup.find_all("a"):
    print("Inner Text: {}".format(link.text))
    print("Title: {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
```
